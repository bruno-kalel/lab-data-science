# 1. Acquire the dataset

# 2. Import all the crucial libraries

```python
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import MinMaxScaler
```

# 3. Import the dataset

```python
dados = pd.read_csv(caminho)
dados
```

```python
dados.columns = dados.columns.str.lower()
```

# 4. Identifying and handling the missing values

```python
dados.isna().sum()
```

```python
dados[dados.isna().any(axis='columns')]
```

```python
dados.salary = SimpleImputer(strategy='mean').fit_transform(
    dados[['salary']]).astype(int)
dados
```

```python
dados.age = SimpleImputer(
    strategy='most_frequent').fit_transform(dados[['age']])
dados
```

```python
dados.isna().sum()
```

# 5. Encoding the categorical data

```python
dados[['country', 'purchased']]
```

```python
dados = pd.get_dummies(dados)
dados
```

# 6. Splitting the dataset

```python
x = dados.drop('purchased_Yes', axis='columns')
y = dados['purchased_Yes']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

# 7. Feature scaling

```python
scaler = MinMaxScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled, columns=x_train.columns)
x_test_scaled = pd.DataFrame(x_test_scaled, columns=x_test.columns)
```
