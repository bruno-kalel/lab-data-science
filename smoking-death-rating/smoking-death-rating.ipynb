# Importando dados e bibliotecas
---

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
```

```python
caminho_2 = ''
```

```python
df_2 = pd.read_csv(caminho_2)
```

```python
df_2.head()
```

```python
df_2 = df_2[df_2['Entity'].isin(['Zimbabwe', 'Puerto Rico', 'United Kingdom',
                                'United States', 'Afghanistan', 'African Region (WHO)', 'Albania'])]
```

```python
df_2.head()
```

```python
df_2.columns = df_2.columns.str.replace(' ', '_').str.lower()
df_2.columns
```

# Removendo a coluna `Code`
---

```python
df_2 = df_2.drop('code', axis='columns')
df_2.head()
```

```python
df_2.info()
```

```python
df_2.smoking_mortality = df_2.smoking_mortality.str.replace(
    '.', '', regex=False).str.strip()
df_2.smoking_mortality = df_2.smoking_mortality.astype('int')
```

# Montando um Dataframe que contenha o total de mortalidade associada ao fumo para cada ano (`Year`), considerando apenas o período entre 1992 e 1998
---

```python
df_questão_6 = df_2[(df_2.year >= 1992) & (df_2.year <= 1998)]

colunas = ['year', 'smoking_mortality']

df_questão_6.groupby('year')['smoking_mortality'].sum()
```

# Montando uma tabela que apresente a **Mediana** da mortalidade associada ao fumo para cada entidade (`Entity`), considerando apenas o período entre 1990 e 1994
---

```python
colunas = ['entity', 'smoking_mortality']

df_2[(df_2.year >= 1990) & (df_2.year <= 1994)][colunas].groupby(
    df_2.entity).median(numeric_only=True)
```

# Apresentando um DataFrame com as entidades (`Entity`) que possuem mortalidade associada ao fumo(`Smoking mortality`) maior que 11.602.391 no ano de 1990
---

```python
df_2[(df_2.smoking_mortality > 11602391) & (df_2.year == 1990)].reset_index()
```

# Encontrando a entidade (`Entity`) com a **menor** e também aquela com a **maior** mortalidade associada ao fumo (**Smoking mortality**), considerando todos os anos (`Year`). Depois, mostrando um dataframe que contenha apenas esses dois registros.
---

```python
colunas = ['entity', 'smoking_mortality']

df_2[colunas].max(), df_2[colunas].min()
```

```python
pd.DataFrame({'máxima mortalidade': df_2[colunas].max(
), 'mínima mortalidade': df_2[colunas].min()}).T
```

# Montando uma tabela que contenha a mortalidade média **POR** entidade (`Entity`) e ano (`Year`) e ordenando (de forma decrescente) os registros encontrados de acordo com o valor do índice de mortalidade associada ao fumo, levando em conta apenas o período entre 1990 e 1994
---

```python
df_questão_10 = df_2[(df_2.year >= 1990) & (df_2.year <= 1994)]
df_questão_10.head()
```

```python
colunas = ['entity', 'year']

df_questão_10.groupby(colunas)[['smoking_mortality']].mean()
```
